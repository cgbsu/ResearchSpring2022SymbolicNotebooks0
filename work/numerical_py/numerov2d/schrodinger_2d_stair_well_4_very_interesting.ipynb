```python
from numerov2d import *
```

```python
pointCount : int = 150
```

```python
xPositions, yPositions = positionGrid(pointCount)
```

```python
potentials : np.ndarray = stairwell(xPositions, yPositions, .01, [1.0 / 3.0, 1.0 / 3.0, 1.0 / 3.0], [1.0 / 3.0, 2.0 / 3.0, 1.0])
```

```python
potentials
```

```python
plt.imshow(potentials)
```

```python
energies, waveFunctions = compute2dWaveFunction(pointCount, potentials, EigenValueTypes.HALF_SPECTRUM)
```

```python
probabilityDistributions = waveFunctions * np.conjugate(waveFunctions)
```

```python
np.max(np.where(waveFunctions[0] != np.conjugate(waveFunctions[0]), 1, 0))
```

```python
decibleNormalized = [10.0 * np.log10(waveFunction * np.conjugate(waveFunction)) for waveFunction in waveFunctions]
```

```python
qfd = np.arange(1, 10)
np.where(qfd > 3, qfd, 0)
```

```python
cut = lambda index, minimum =  50 : np.where((decibleNormalized[index] > -minimum), decibleNormalized[index], -minimum)
```

```python
plt.figure(figsize=(9, 9))
plt.contourf(xPositions, yPositions, cut(0), 200)
plt.colorbar()
```

```python
plt.figure(figsize=(9, 9))
plt.contourf(xPositions, yPositions, cut(3), 200)
plt.colorbar()
```
